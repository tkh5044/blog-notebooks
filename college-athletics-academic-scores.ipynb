```python
teams = ncaa.groupby(['SCHOOL_NAME', 'SPORT_NAME'])[cols].mean().sort_values('2014_SCORE', ascending=False).reset_index()
```

```python
teams['avg'] = (teams['2014_SCORE'] + teams['2013_SCORE'] + teams['2012_SCORE'] + teams['2011_SCORE'] + 
                teams['2010_SCORE'] + teams['2009_SCORE'] + teams['2008_SCORE'] + teams['2007_SCORE'] +
                teams['2006_SCORE'] + teams['2005_SCORE'] + teams['2004_SCORE']) / 11
```

```python
teams[teams.avg == 1000][['SCHOOL_NAME', 'SPORT_NAME', 'avg']].sort_values('avg', ascending=False)
```

```python
team1000 = teams[teams.avg == 1000][['SCHOOL_NAME', 'SPORT_NAME', 'avg']].sort_values('avg', ascending=False)
```
