```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
import re
```

```python
def get_lyrics(album_name, songs, filename):
    lyrics = []
    for i in songs:
        url = 'https://genius.com/Kendrick-lamar-' + i + '-lyrics'
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        lyrics.append(soup.find('p').text)
        sleep(1)
        
    df = pd.DataFrame(columns=['album', 'song', 'lyrics'])
    df.song = songs
    df.lyrics = lyrics
    df.album = album_name
    df.lyrics = df.lyrics.str.encode('utf-8').astype(str)
    
    new_lyrics = []
    for i in df.lyrics:
        new_lyrics.append(re.sub('\[.+]', '', i))
    df.lyrics = new_lyrics
    df.lyrics = df.lyrics.str.replace('\n', ' ').str.replace("'", '')
    
    df.to_csv(filename + '.csv', index=False, encoding='utf-8')
```

# Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
cv.fit(damn.lyrics)
len(cv.get_feature_names())
```

```python
damn_words = pd.DataFrame(cv.transform(damn.lyrics).todense(), columns=cv.get_feature_names())
```

```python
word_counts = damn_words.sum(axis=0).sort_values(ascending = False)
word_counts.head(20)
```

```python
# top 10 most used words for each song

for i, j in enumerate(damn_songs):
    print j.upper() + '.'
    print damn_words.transpose().sort_values(i, ascending=False)[i].head(10).transpose()
    print
```
